# Primer notebook en Google Colab
Este notebook tiene la finalidad de familiarizarse con la herramienta, es un entregable para la clase de Programacion con Python.
Como actividad en este notebook mostrare dos graficas utilizando la libreria matplotlib.


```python
# Se importan las librerías

import numpy as np
import matplotlib.pyplot as plt

""" Se define el rango de la gráfica en el eje de las X,
en este caso 20 números"""

X = range(1, 21)

"""Se definen los valores del eje de las Y, aquí es donde colocamos 
nuestra ecuación. Por ejemplo, 2x + 1, la cual es una función lineal
'Y' recibe un arreglo de números, porque 
estamos multiplicando los valores en X"""

Y = [(2 * value) + 1 for value in X]
print(Y)

"""Se le definen los valores a la gráfica 
con las variables X y Y anteriores"""

plt.plot(X,Y)

# Se define el nombre que tendrá el eje de las X

plt.xlabel('Eje de X')

# Se define el nombre que tendrá el eje de las Y

plt.ylabel('Eje de Y')

# Se define el título de la grafica

plt.title('Grafica función lineal')
# Se despliega la grafica

plt.show()

"""Ahora haremos el mismo procedimiento para una función cuadrática
El procedimiento es el mismo, solo cambiamos nuestra ecuación"""

# Se define el rango de la gráfica en el eje X, en este caso 20 números

X = range(1, 21)

"""Se define los valores del eje Y, aquí se aplica la ecuación
Por ejemplo, 2x^2 - 4x - 1, la cual es una función lineal
'Y' recibe un array de números, se está multiplicando los valores en X"""

Y = [(2 * value**2) - (4 * value) - 1 for value in X]
print(Y)

# Se define el valor de los ejes X y Y a la gráfica

plt.plot(X,Y)

# Se define el nombre que tendrá el eje de las X

plt.xlabel('Eje de X')

# Se define el nombre que tendrá el eje de las Y

plt.ylabel('Eje de Y')

# Se define el título de la grafica

plt.title('Grafica función cuadrática')

# Se despliega la grafica

plt.show()
```
